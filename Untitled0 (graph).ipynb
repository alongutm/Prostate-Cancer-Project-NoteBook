defining data types for objects:

```
def convert_object_to_date(df):
  date_columns = ['DATE OF BIRTH','DATE OF DIAGNOSIS (BIOPSY)',
                  'RT START', 'RT END' ,'HT START','HT END',
                  'psa nadir DATE','BF DATE','Date of LF',
                  'Date of DF','Date of first event']
  for column in date_columns:
    df[column] = pd.to_datetime(df[column])
    df[column] = (df[column]-df[column].min())/ np.timedelta64(1,'D')

def convert_gleason_score(df):
  df['GLEASON SCORE'] = df['GLEASON SCORE'].apply(lambda x : int(x[0]))

convert_object_to_date(df_belinson)
convert_gleason_score(df_belinson)
df_belinson.info()
```

definining null handaling methods:

```
def fill_t_stage_missing_values(df):
  commun = df['T STAGE (* = AT LEAST)'].mode()[0]
  df['T STAGE (* = AT LEAST)'].fillna(inplace = True, value = commun)

def fill_pni(df):
  commun = df['PNI'].mode()[0]
  df['PNI'].fillna(inplace = True, value = commun)

def fill_sv(df):
  commun = df['SV'].mode()[0]
  df['SV'].fillna(inplace = True, value = commun)

def fill_ht_duration(df):
  df['HT DURATION'].fillna(inplace = True, value=1)

def fill_ht_type(df):
  df['HT TYPE'].fillna(inplace = True, value='UNKNONW')

def fill_psa_nadir(df):
  avg = np.nanmean(df['psa nadir PSA'])
  df['psa nadir PSA'].fillna(inplace = True, value = avg)
  commun = df['psa nadir DATE'].mode()[0]
  df['psa nadir DATE'].fillna(inplace=True, value = commun)

def fill_bf(df):
  df['PSA AT BF'].fillna(inplace = True, value = 0)
  df['NADIR-BF INTERVAL (mo)'].fillna(inplace = True, value = 0)
  df['BF DATE'].fillna(inplace= True, value = np.datetime64('1900-01-01'))
  commun = df['BF'].mode()[0]
  df['BF'].fillna(inplace = True, value = commun)

def fill_df(df):
  df['Time to DF, months'].fillna(inplace = True, value = 0)

def fill_time_of_deat(df):
  df['TIME TO DEATH (mo)'].fillna(inplace = True, value = 0)

def fill_time_to_lf(df):
  df['Time to LF, months.1'].fillna(inplace = True, value = 0)

fill_t_stage_missing_values(df_belinson)
fill_pni(df_belinson)
fill_sv(df_belinson)
fill_ht_duration(df_belinson)
fill_ht_type(df_belinson)
fill_psa_nadir(df_belinson)
fill_bf(df_belinson)
fill_df(df_belinson)
fill_time_of_deat(df_belinson)
fill_time_to_lf(df_belinson)
```

```
def drop_dates(df):
  date_columns = ['DATE OF BIRTH','DATE OF DIAGNOSIS (BIOPSY)',
                  'RT START', 'RT END' ,'HT START','HT END',
                  'psa nadir DATE','BF DATE','Date of LF',
                  'Date of DF','Date of first event']
  for column in date_columns:
    try:
      df.drop(column,  axis='columns', inplace=True)
    except:
      print(column, " not exist")
drop_dates(df_belinson)
```
